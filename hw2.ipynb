## Merge sorted stream

```python
def merge_2sorted_stream(stream1, stream2):
    """
    function to merge two sorted streams
    here we assume the input stream may
    be infinite, so we don't access the
    element using index nor call __len__
    """
    try:
        x = next(stream1)
    except StopIteration:
        x = None

    try:
        y = next(stream2)
    except StopIteration:
        y = None

    # implement merge sort here
    while (x != None) and (y != None):
        if (x < y):
            yield x
            try:
                x = next(stream1)
            except StopIteration:
                x = None
        else:
            yield y
            try:
                y = next(stream2)
            except StopIteration:
                y = None

    # deal with the edge cases
    if (x == None) and (y == None):
        yield None
    elif (x == None) and (y != None):
        yield y
        for i in stream2:
            yield i
    elif (y == None) and (x != None):
        yield x
        for i in stream1:
            yield i


def merge_sorted_stream(*args):
    """
    We use recursion technique to merge
    varying number of sorted streams by
    making use of our 2-stream version
    merge function

    merge(s1, s2, s3, ...) = merge(merge(s1, s2), s3, ...)

    """
    if len(args) <= 1:
        print("please provide at least two streams")
        return
    else:
        g = merge_2sorted_stream(args[0], args[1])
        if(len(args) == 2):
            for x in g:
                yield x
            return
        else:
            for x in merge_sorted_stream(g, *args[2:]):
                yield x
```

## Tree traversal

```python
class TreeNode:

    def __init__(self, x):
        self.info = str(x)
        self.left = None
        self.right = None

    def in_order(self):
        # Inorder (Left, Root, Right)
        if self.left != None:
            for i in self.left.in_order():
                yield i

        yield self.info

        if self.right != None:
            for i in self.right.in_order():
                yield i

    def pre_order(self):
        # Preorder (Root, Left, Right)
        yield self.info

        if self.left != None:
            for i in self.left.pre_order():
                yield i

        if self.right != None:
            for i in self.right.pre_order():
                yield i

    def post_order(self):
        # Postorder (Left, Right, Root)
        if self.left != None:
            for i in self.left.post_order():
                yield i

        if self.right != None:
            for i in self.right.post_order():
                yield i

        yield self.info


root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

print(' -> '.join(item for item in root.in_order()))
print(' -> '.join(item for item in root.pre_order()))
print(' -> '.join(item for item in root.post_order()))
```

## Implement a timer

```python
import time
from contextlib import ContextDecorator


class timer(ContextDecorator):

    def __init__(self):
        self.start = 0
        self.end = 0

    def __enter__(self):
        self.start = time.time()
        return self

    def __exit__(self, *exc):
        self.end = time.time()
        print("Total execution time: %lf seconds" % (self.end-self.start))
        return False


with timer() as timer:
    time.sleep(1)


@timer
def sleep(secs):
    time.sleep(secs)


sleep(1)
```
